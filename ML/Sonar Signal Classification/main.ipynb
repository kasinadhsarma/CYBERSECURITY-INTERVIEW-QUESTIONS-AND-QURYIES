Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Import Datasets

```python
sonar_data = pd.read_csv('/content/Copy of sonar data.csv', header = None)
```

Data pre-processing techniques

```python
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
x = sonar_data.drop(columns = 60)
y = sonar_data[60]
```

```python
print(x)
```

```python
print(y)
```

Splitting the Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 3)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(x_train)
print(y_train)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

Predicting Our Original Data

```python
x_train_prediction = model.predict(x_train)
```

```python
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on training Data is', training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
```

```python
training_test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy on testing Data is', training_test_data_accuracy)
```
